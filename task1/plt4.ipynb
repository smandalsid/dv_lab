```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv('/home/siddharth/Documents/ubuntu_python_work/pandas/olympic_medals.csv')
```

```python
years=[]
for i in range(len(df['slug_game'])):
    if(int(df['slug_game'][i][-4:]) not in years):
        years.append(int(df['slug_game'][i][-4:]))
```

```python
us_medals=[]
cn_medals=[]
for i in range(len(years)):
    us_medals.append(0)
    cn_medals.append(0)
for i in range(len(df['slug_game'])):
    if(int(df['slug_game'][i][-4:]) in years and df['country_code'][i]=='US'):
        us_medals[years.index(int(df['slug_game'][i][-4:]))]+=1
    if(int(df['slug_game'][i][-4:]) in years and df['country_code'][i]=='CN'):
        cn_medals[years.index(int(df['slug_game'][i][-4:]))]+=1
```

```python
us_medals=np.array(us_medals)
cn_medals=np.array(cn_medals)
years=np.array(years)
```

```python
plt.scatter(years, us_medals, color="green")
plt.scatter(years, cn_medals, color="#88c999")
plt.legend(['USA', 'CHINA'])
plt.xlabel('Year', fontsize="15", color="olivedrab")
plt.ylabel('Number of medals won', fontsize="15", color="olivedrab")
plt.title("Comparison of China vs USA in the olympics", fontsize="20", color="forestgreen")
plt.show()
```
