```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
df=pd.read_csv('/home/siddharth/Documents/ubuntu_python_work/pandas/olympic_medals.csv')
```

```python
df.dtypes
```

```python
medals=['GOLD', 'SILVER', 'BRONZE']
medal_count=[0,0,0]
```

```python
for i in range(len(df['medal_type'])):
    if(df['country_name'][i]=='India'):
        medal_count[medals.index(df['medal_type'][i])]+=1
```

```python
for i in range(len(medals)):
    medals[i]+='-'
    medals[i]+=str(medal_count[i])
```

```python
plt.pie(medal_count, labels=medals)
plt.title("Share of type of medals won by India")
plt.legend()
```

```python
male=0
female=0
mixed=0
for i in range(len(df['medal_type'])):
    if df['country_name'][i]=='India':
        if df['event_gender'][i]=='Men':
            male+=1
        elif df['event_gender'][i]=='Women':
            female+=1
        else:
            mixed+=1
```

```python
medal_gender_dist=[male, female, mixed]
plt.pie(medal_gender_dist, labels = ['Men'+'-'+str(male), 'Women'+'-'+str(female), 'Mixed'+'-'+str(mixed)])
plt.title("Share of men and women in the total number of medals")
plt.legend()
```
