```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv('/home/siddharth/Documents/ubuntu_python_work/pandas/olympic_medals.csv')
```

```python
d_gold={}

for i in range(len(df['country_code'])):
    if(df['medal_type'][i]=='GOLD'):
        if(df['country_code'][i] not in d_gold):
            d_gold[df['country_code'][i]]=0
        else:
            d_gold[df['country_code'][i]]+=1
```

```python
d_silver={}

for i in range(len(df['country_code'])):
    if(df['medal_type'][i]=='SILVER'):
        if(df['country_code'][i] not in d_silver):
            d_silver[df['country_code'][i]]=0
        else:
            d_silver[df['country_code'][i]]+=1
```

```python
d_bronze={}

for i in range(len(df['country_code'])):
    if(df['medal_type'][i]=='BRONZE'):
        if(df['country_code'][i] not in d_bronze):
            d_bronze[df['country_code'][i]]=0
        else:
            d_bronze[df['country_code'][i]]+=1
```

```python
x=[]
num_bins=25
for i in d_gold:
    x.append(d_gold[i])
plt.hist(x, num_bins, facecolor='blue', alpha=0.5, label='GOLD')

y=[]
for i in d_silver:
    y.append(d_silver[i])
plt.hist(y, num_bins, facecolor='green', alpha=0.5, label='SILVER')

z=[]
for i in d_bronze:
    z.append(d_bronze[i])
plt.hist(z, num_bins, facecolor='red', alpha=0.5, label='BRONZE')
plt.xlabel('Number of medals')
plt.ylabel('Number of countries')
plt.title("Representation of number of countries with the number of medals they have won")
plt.legend()
plt.show()
```
