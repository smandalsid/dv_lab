```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import matplotlib.colors as clr
```

```python
df=pd.read_csv("olympic_medals.csv")
```

```python
medals1=[]
medals2=[]
sports=[]

for i in range(len(df)):
    if len(sports)!=3:
        if df["discipline_title"][i] not in sports:
            sports.append(df["discipline_title"][i])
            medals1.append(0)
            medals2.append(0)
    else:
        break

for i in range(len(df)):
    if df["country_code"][i]=="US" and df["discipline_title"][i] in sports:
        medals1[sports.index(df["discipline_title"][i])]+=1

for i in range(len(df)):
    if df["country_code"][i]=="CN" and df["discipline_title"][i] in sports:
        medals2[sports.index(df["discipline_title"][i])]+=1
```

```python
fig,(axes1,axes2)=plt.subplots(1,2)
axes1.set_title("USA")
axes1.set_xlabel("Medals")
axes1.set_ylabel("Sports")
axes2.set_title("CHINA")
axes2.set_xlabel("Medals")
axes2.set_ylabel("Sports")

axes1.scatter(sports,medals1,marker='X',s=140,c="red")
fig.legend(loc=8,framealpha=1, fontsize=8)
axes2.scatter(sports,medals2,marker='s',s=140,c="blue")

plt.rcParams["figure.figsize"] = [20, 10]

plt.show()
```
