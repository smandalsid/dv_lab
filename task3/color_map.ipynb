```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import matplotlib.colors as clr
```

```python
df=pd.read_csv("olympic_medals.csv")
```

```python
medals=[]
sports=[]

for i in range(len(df)):
    if len(sports)!=12:
        if df["discipline_title"][i] not in sports:
            sports.append(df["discipline_title"][i])
            medals.append(0)
    else:
        break

for i in range(len(df)):
    if df["country_code"][i]=="US" and df["discipline_title"][i] in sports:
        medals[sports.index(df["discipline_title"][i])]+=1

```

```python
z=np.linspace(-1, 1, 12)
colors=['red', 'orange', 'khaki', 'yellow', 'greenyellow', 'green', 'mediumseagreen', 'turquoise', 'cyan', 'navy', 'darkorchid', 'magenta']

cmap=clr.ListedColormap(colors)
figure, axes = plt.subplots(ncols=1)
scatter=axes.scatter(sports, medals, s=150, c=z, cmap=cmap)
figure.colorbar(scatter)
plt.rcParams["figure.figsize"] = [20, 10]
axes.set_title("Medals won by USA in 12 sports")
plt.show()
```
